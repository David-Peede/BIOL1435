## Drift

```python
# Intialize the number of replicate simulations.
reps = 100
# Intialize the population size.
N = 100
# Intialize the starting frequency.
p = 0.5
# Intialize the number of generations to simulate.
gens = 100
# Intialize a random seed for reporducibility.
np.random.seed(42)
# Intailize a matrix to store the simulated results.
freq_mat = np.zeros((reps, gens))
# Set the first generation to the starting frequency.
freq_mat[:, 0] = p
# For every replicate...
for rep in range(reps):
    # For every generation...
    for gen in range(gens - 1):
        # Sample two alleles per diploid sample.
        genotypes = np.random.binomial(n=2, p=freq_mat[rep, gen], size=2*N)
        # Determine the allele frequency in the current generation.
        freq = genotypes.sum() / (genotypes.size * 2)
        # Update the results matrix.
        freq_mat[rep, gen + 1] = freq
```

```python
# Intialize the figure.
fig = plt.figure(constrained_layout=True, dpi=300.0)
# Intialize the axis.
ax = fig.add_subplot(111)
# Plot the allele frequency trajectories.
ax.plot(freq_mat.T)
# Plot the mean.
ax.plot(
    freq_mat.mean(axis=0), color='black',
    linestyle='dashed', linewidth=3, label='Mean',
)
# Title the plot.
ax.set_title(r'$\mathbf{p_0}$ = 0.5')
# Label the axes.
ax.set_xlabel('Time (generations)')
ax.set_ylabel(r'Allele Frequency ($\mathbf{p}$)')
# Label the axes.
ax.set_xticks([1, 50, 100])
ax.set_yticks(np.linspace(0, 1, 5))
ax.set_xticklabels(['1', '50', '100'])
ax.set_yticklabels(np.linspace(0, 1, 5).astype(str))
# Plot the legend.
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5), frameon=False)
# Save as a svg.
plt.savefig('../figures/6apr23_drift.svg', dpi=300, format='svg')
# Show the plot.
plt.show()
```

## Selection

```python
# Define a function to calculate the mean fitness of a population.
def wbar(p, w11, w12, w22):
    # Calculate q.
    q = 1 - p 
    # Calcalte the average fitness in the population.
    wbar = (p**2 * w11) + (2 * p * q * w12) + (q**2 * w22)
    return wbar
```

```python
# Determine the allele frequency in the next generation prior to selection.
def p_prime(p, w11, w12, w22):
    # Calculate q.
    q = 1 - p 
    # Calcalte the average fitness in the population.
    w_bar = wbar(p, w11, w12, w22)
    # Calculate the numerator of p prime.
    p_prime_num = p * ((p * w11) + (q * w12))
    # Compute the allele  frequency before selection in the next generation.
    p_prime = p_prime_num / w_bar
    return p_prime    
```

```python
# Intialize the starting frequency.
p0 = 0.5
# Intialize the number of generations to simulate.
gens = 100
# Intialize relative fitness.
w_11 = 1
w_12 = 1
w_22 = 1
# Intialize a random seed for reporducibility.
np.random.seed(42)
# Intailize a matrix to store the simulated results.
freq_array = np.zeros(gens)
# Set the first generation to the starting frequency.
freq_array[0] = p0
# For every generation...
for gen in range(gens - 1):
    # Update the results matrix.
    freq_array[gen + 1] = p_prime(p=freq_array[gen], w11=w_11, w12=w_12, w22=w_22)
```

```python
# Intialize the figure.
fig = plt.figure(constrained_layout=True, dpi=300.0)
# Intialize the axis.
ax = fig.add_subplot(111)
# Plot the allele frequency trajectories.
ax.plot(freq_array, linewidth=3)
# Title the plot.
ax.set_title(r'$\mathbf{p_0}$ = 0.5; $\mathbf{w_{11} = w_{12} = w_{22}}$')
# Label the axes.
ax.set_xlabel('Time (generations)')
ax.set_ylabel(r'Allele Frequency ($\mathbf{p}$)')
# Label the axes.
ax.set_xticks([1, 50, 100])
ax.set_yticks(np.linspace(0, 1, 5))
ax.set_xticklabels(['1', '50', '100'])
ax.set_yticklabels(np.linspace(0, 1, 5).astype(str))
# Save as a svg.
plt.savefig('../figures/6apr23_p_prime_no_sel.svg', dpi=300, format='svg')
# Show the plot.
plt.show()
```

```python
# Define a function to calculate the difference in allele frequencies due to selection.
def delta_p(p, w11, w12, w22):
    # Calculate q.
    q = 1 - p 
    # Calcalte the average fitness in the population.
    w_bar = wbar(p, w11, w12, w22)
    # Calculate the numerator of delta p.
    delta_p_num = (p * q) * ((p * (w11 - w12)) + (q * (w12 - w22)))
    # Compute the difference in allele frequencies due to selection.
    delta_p = delta_p_num / w_bar
    return delta_p    
```

```python
# Intialize the starting frequency.
p0 = 0.5
# Intialize the number of generations to simulate.
gens = 100
# Intialize relative fitness.
w_11 = 1
w_12 = 1
w_22 = 1
# Intialize a random seed for reporducibility.
np.random.seed(42)
# Intailize a matrix to store the simulated results.
freq_array = np.zeros(gens)
# Set the first generation to the starting frequency.
freq_array[0] = p0
# For every generation...
for gen in range(gens - 1):
    # Update the results matrix.
    freq_array[gen + 1] = delta_p(p=freq_array[gen], w11=w_11, w12=w_12, w22=w_22)
```

```python
# Intialize the figure.
fig = plt.figure(constrained_layout=True, dpi=300.0)
# Intialize the axis.
ax = fig.add_subplot(111)
# Plot the allele frequency trajectories.
ax.plot(freq_array, linewidth=3)
# Title the plot.
ax.set_title(r'$\mathbf{p_0}$ = 0.5; $\mathbf{w_{11} = w_{12} = w_{22}}$')
# Label the axes.
ax.set_xlabel('Time (generations)')
ax.set_ylabel(r'$\mathbf{\Delta p}$')
# Label the axes.
ax.set_xticks([1, 50, 100])
ax.set_yticks(np.linspace(0, 1, 5))
ax.set_xticklabels(['1', '50', '100'])
ax.set_yticklabels(np.linspace(0, 1, 5).astype(str))
# Save as a svg.
plt.savefig('../figures/6apr23_delta_p_no_sel.svg', dpi=300, format='svg')
# Show the plot.
plt.show()
```

## Directional Selection

```python
# Determine the allele frequency in the next generation prior to selection.
def p_prime(p, h, s):
    # Intialize the relative fitness.
    w11 = 1
    w12 = 1 - (h * s)
    w22 = 1 - s
    # Calculate q.
    q = 1 - p 
    # Calcalte the average fitness in the population.
    w_bar = wbar(p, w11, w12, w22)
    # Calculate the numerator of p prime.
    p_prime_num = p * ((p * w11) + (q * w12))
    # Compute the allele  frequency before selection in the next generation.
    p_prime = p_prime_num / w_bar
    return p_prime    
```

```python
# Intialize the starting frequency.
p0 = 0.01
# Intialize the number of generations to simulate.
gens = 5_000
# Intialize a random seed for reporducibility.
np.random.seed(42)
# Intailize a matrix to store the simulated results.
h0_array = np.zeros(gens)
h05_array = np.zeros(gens)
h1_array = np.zeros(gens)
# Set the first generation to the starting frequency.
h0_array[0] = p0
h05_array[0] = p0
h1_array[0] = p0
# For every generation...
for gen in range(gens - 1):
    # Update the results matrix.
    h0_array[gen + 1] = p_prime(p=h0_array[gen], h=0, s=0.01)
    h05_array[gen + 1] = p_prime(p=h05_array[gen], h=0.5, s=0.01)
    h1_array[gen + 1] = p_prime(p=h1_array[gen], h=1, s=0.01)
```

```python
# Intialize the figure.
fig = plt.figure(constrained_layout=True, dpi=300.0)
# Intialize the axis.
ax = fig.add_subplot(111)
# Plot the allele frequency trajectories.
ax.plot(h0_array, label=r'$h = 0$')
ax.plot(h05_array, label=r'$h = 0.5$')
ax.plot(h1_array, label=r'$h = 1$')
# Title the plot.
ax.set_title(r'$\mathbf{p_0}$ = 0.01; $\mathbf{s = 0.01}$')
# Label the axes.
ax.set_xlabel('Time (generations)')
ax.set_ylabel(r'Allele Frequency ($\mathbf{p}$)')
# Label the axes.
ax.set_xticks(np.arange(0, 5_000+1, 1_000))
ax.set_yticks(np.linspace(0, 1, 5))
ax.set_xticklabels(np.arange(0, 5_000+1, 1_000).astype(str))
ax.set_yticklabels(np.linspace(0, 1, 5).astype(str))
# Plot the legend.
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5), frameon=False)
# Save as a svg.
plt.savefig('../figures/6apr23_directional_sel.svg', dpi=300, format='svg')
# Show the plot.
plt.show()
```
