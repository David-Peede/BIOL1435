```python
# Import modules.
import demes
import demesdraw
import numpy as np
import stdpopsim
import msprime
from matplotlib import pyplot as plt
import toyplot
import toyplot.svg
import toytree
```

## Population vs Differentiation vs Structure

```python
# Intialize the species information.
species = stdpopsim.get_species('AnaPla')
# Intialize the demographic model.
demo = species.get_demographic_model('MallardBlackDuck_2L19')
# Convert the demographic model to a demes graph.
graph = msprime.Demography.to_demes(demo.model)
# Update the matplolib style to mimic the toyplot style.
plt.rcParams.update({
    'figure.figsize': (toyplot.units.convert('500px', 'in'), toyplot.units.convert('325px', 'in')),
    'font.family': 'sans-serif',
    'font.sans-serif': 'Helvetica',
    'font.size': toyplot.units.convert('9px', 'pt'),
    'axes.titlesize': toyplot.units.convert('14px', 'pt'),
    'axes.titleweight': 'bold',
    'axes.labelsize': toyplot.units.convert('12px', 'pt'),
    'axes.labelweight': 'bold',
})
# Intialize the figure.
fig = plt.figure(constrained_layout=True, dpi=300.0)
# Intialize the axes.
ax = fig.add_subplot(111)
# Plot the demes graph.
demesdraw.tubes(
    graph, ax=ax,
    log_time=demesdraw.utils.log_time_heuristic(graph),
    title='North American Duck Demography (Lavretsky et al., 2020)', seed=42,
)
# Save as a svg.
plt.savefig('../figures/23feb23_duck_demo.svg', dpi=300, format='svg')
# Show the plot.
plt.show()
```

```python
# Simulate 5 samples per population.
ts = msprime.sim_ancestry(
    samples=[
        msprime.SampleSet(3, ploidy=1, population='Mallard'),
        msprime.SampleSet(3, ploidy=1, population='Black_duck'),
    ],
    demography=demo.model,
    random_seed=21,
)
# Extract the first tree and convert to newick format.
newick = ts.first().newick(
    node_labels= {
        0: 'Mallard 1', 1: 'Mallard 2', 2: 'Mallard 3',
        3: 'Black Duck 1', 4: 'Black Duck 2', 5: 'Black Duck 3',
        
    }
)
tree = toytree.tree(newick, tree_format=0)
# Adjust the edge thickness.
e_widths = [5 for i in tree.get_edge_values('idx')]
# Intialize the canvas.
canvas = toyplot.Canvas(width=700, height=500)
# Intialize the axes.
axes = canvas.cartesian()
# Plot the tree.
mark = tree.draw(
    ts='s',
    edge_widths=e_widths,
    node_sizes=15,
    node_labels=False,
    tip_labels=True,
    scalebar=False,
    axes=axes,
);
axes.show = False
```

```python
# Save as a svg.
toyplot.svg.render(canvas, '../figures/23feb23_duck_structure.svg')
```

## Limitations of $F_{ST}$

```python
# Intialize the efffective population size.
Ne = 1e3
# Intialize the length of simulated region.
L = 1e6
# Intialize the number of simulated replicates.
num_reps = 100
# Define the slective sweep model.
sweep_model = msprime.SweepGenicSelection(
    position=L / 2,  # Location of the beneficial mutation.
    start_frequency=1.0 / (2 * Ne), 
    end_frequency=1.0 - (1.0 / (2 * Ne)),
    s=0.25, # Selection coefficient of the benefecial mutation.
    dt=1e-6,
)
# Simulate a selective sweep that occurs in the immediate past,
# followed by the neutral coalescent for the rest of history.
reps = msprime.sim_ancestry(
    5,
    model=[sweep_model, msprime.StandardCoalescent()],
    population_size=Ne,
    recombination_rate=1e-7,
    sequence_length=L,
    num_replicates=num_reps,
    random_seed=42,
)
# Create start and end points for 50kb windows.
wins = np.linspace(0, L, 21)
# Determine the midpoint of each window.
mids = (wins[1:] + wins[:-1]) / 2
# Intialize a results matrix.
diversity = np.zeros((num_reps, mids.size))
# For every replicate simulation...
for rep, ts in enumerate(reps):
    # Estimate diversity (aka pi).
    diversity[rep] = ts.diversity(windows=wins, mode='branch')
```

```python
# Intialize a canvas.
canvas = toyplot.Canvas(
    width=500, height=325,
)
# Intialize and label the axes.
axes = canvas.cartesian(
    xlabel='Position (bp)',
)

# Plot the neutral expectation.
avg_fst = axes.plot(
    mids, np.full((mids.size), 0.5),
    style={'stroke-dasharray': 5},
)
# Plot the pi at the selective sweep.
sweep_pi = axes.plot(
    mids,
    (diversity.mean(axis=0) / (diversity.mean(axis=0).max()+2000)),
)
# Plot fst at the selective sweep.
sweep_fst = axes.plot(
    mids,
    np.abs(1.05 - (diversity.mean(axis=0) / (diversity.mean(axis=0).max()+2000))),
)
# Plot the legend.
canvas.legend(
    [
        ('<i>F<sub>ST</sub></i> (Genome Average)', avg_fst),
        (u'\u03c0 (Sweep)', sweep_pi),
        ('<i>F<sub>ST</sub></i> (Sweep)', sweep_fst),
    ],
    corner=('top-left', 50, 50, 50),
);
```

```python
# Save as a svg.
toyplot.svg.render(canvas, '../figures/23feb23_rel_diff.svg')
```

## Advantages of $d_{XY}$

```python
# Create a newick format with two trees.
newick = """\
((Mallard 1:1.5, Mallard 2:1.5):2,(Black Duck 1:1.5, Black Duck 2:1.5):2);
((Mallard 1:0.5, Mallard 2:0.5):2,(Black Duck 1:0.5, Black Duck 2:0.5):2);
"""
# Read in the multi-tree object.
mtree = toytree.mtree(newick)
# Intialize the canvas.
canvas = toyplot.Canvas(width=700, height=500)
# Intialize axes.
ax1= canvas.cartesian(grid=(2, 1, 0), xshow=False, yshow=False)
ax2 = canvas.cartesian(grid=(2, 1, 1), xshow=False, yshow=False)
# Adjust the edge thickness.
e1_widths = [5 for i in mtree.treelist[0].get_edge_values('idx')]
e2_widths = [5 for i in mtree.treelist[1].get_edge_values('idx')]
# Plot the trees.
mtree.treelist[0].draw(
    ts='o',
    edge_widths=e1_widths,
    node_sizes=15,
    node_labels=False,
    tip_labels=True,
    scalebar=True,
    axes=ax1,
)
mtree.treelist[1].draw(
    ts='o',
    edge_widths=e2_widths,
    node_sizes=15,
    node_labels=False,
    tip_labels=True,
    scalebar=False,
    axes=ax2,
)

# Remove tick labels.
ax1.x.ticks.labels.show = False
ax1.x.ticks.show = False
# Label the y.
ax1.x.label.text = '<i>d<sub>XY</sub></i>                     '
```

```python
# Save as a svg.
toyplot.svg.render(canvas, '../figures/23feb23_abs_diff.svg')
```
